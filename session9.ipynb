```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.uniform(0,1,1000)

plt.hist(random_numbers, bins=100, color='skyblue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers Uniformly Distributed in [0,1]')

plt.savefig('random_numbers_histogram.pdf', format='pdf')
```
